```python
import numpy as np
```

```python
X=np.array(([2,9],[1,5],[3,6]),dtype=float)
y=np.array(([92],[86],[89]),dtype=float)
print("X= ",X)
print("y= ",y)
```

Normalising the vectors

```python
X=X/np.amax(X,axis=0)
y=y/100
print("X= ",X)
print("y= ",y)
```

Defining Sigmoid function

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

Defining Derivative of Sigmoid function

```python
def derivatives_sigmoid(x):
    return x*(1-x)
```

Initialising the layer parameters

```python
epoch=7000
lr=0.1
inputlayer_neurons=2
hiddenlayer_neurons=3
output_neurons=1
```

```python
#initialising weights between input layer and hidden layer
wh=np.random.uniform(size=(inputlayer_neurons,hiddenlayer_neurons))
#initialising bias between input layer and hidden layer
bh=np.random.uniform(size=(1,hiddenlayer_neurons))
#initialising weights between hidden layer and output layer
wout=np.random.uniform(size=(hiddenlayer_neurons,output_neurons))
#initialising bias between hidden layer and output layer
bout=np.random.uniform(size=(1,output_neurons))
print("wh= ",wh)
print("bh= ",bh)
print("wout= ",wout)
print("bout= ",bout)
```

```python
for i in range(epoch):
      #dot product of inputs and weights
      hinp1=np.dot(X,wh)
      print("hinp1= ",hinp1)
      #adding bias
      hinp=hinp1+bh
      print("hinp= ",hinp)
      #activation
      hlayer_act=sigmoid(hinp)
      #dot product of hidden layer input and weights
      outinp1=np.dot(hlayer_act,wout)
      #adding bias
      outinp=outinp1+bout
      output=sigmoid(outinp)
      print("Actual output: ",output)
      print("Expected output: ",y)
      print("Propagating back the error...")
      EO=y-output
      outgrad=derivatives_sigmoid(output)
      d_output=EO*outgrad
      EH=d_output.dot(wout.T)
      hiddengrad=derivatives_sigmoid(hlayer_act)
      d_hiddenlayer=EH*hiddengrad
      wout+=hlayer_act.T.dot(d_output)*lr
      wh+=X.T.dot(d_hiddenlayer)*lr
      print("Input:\n"+str(X))
      print("Actual Output:/n"+str(y))
      print("Predicted Output:/n",output)
```
